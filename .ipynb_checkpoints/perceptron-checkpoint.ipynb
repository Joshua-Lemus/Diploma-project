```python
import numpy as np
import matplotlib.pyplot as plt

import math
from cubature import cubature
import scipy.optimize as optimize
from scipy.stats import multivariate_normal
from scipy.special import erf, erfc
```

## Perceptron learning rule

```python
N = 100
alpha = 1
P = int(alpha * N)
learn_bias = False

X = np.random.randn(P, N)
y = np.random.randint(2, size=P)

if learn_bias:
    X = np.c_[X, np.ones((P,1))]
```

### Sequential implementation

```python
# set parameters
lr = 0.1
num_epochs = 1000
print_every = 100

# init weight vector
w = np.random.randn(N) / np.sqrt(N)
w = np.zeros(N)
if learn_bias:
    w = np.r_[w, np.random.randn()]
    
w0 = w.copy()
        
# compute initial error
errs = []
S = 1. * ((X @ w) > 0)    # This is our prediction for y (heaviside(X W))
err = np.abs(S - y).sum() # Score error
errs += [err]

# run training                      # Gradient descent, the error function is 1/2 \sum_i (y_i-s_i)^2
for ep in range(num_epochs):
    perm = np.random.permutation(P)
    for mu in perm:
        s = 1. * ((w @ X[mu]) > 0)  # same as S, but for this datapoint
        delta = y[mu] - s           # y - theta(X W)
        if delta != 0:              
            w += lr * delta * X[mu] # The gradient is (y - theta(X W)) X
    S = 1. * ((X @ w) > 0)       # Re-estimate error
    err = np.abs(S - y).sum()    
    errs += [err]                # Keep track of errors
    
    if ep % print_every == 0:
        print(f"ep: {ep} err: {err}")
    
    if err == 0:
        print("solution found -> EXITING!")
        break
        
plt.plot(errs, '.-');
plt.ylabel("errors");
plt.xscale('log');
plt.yscale('log');
```

### parallel implementation

```python
# set parameters
lr = 0.01
num_epochs = 1000
print_every = 10

# init weight vector
w = np.random.randn(N) / np.sqrt(N)
if learn_bias:
    w = np.r_[w, np.random.randn()]
# w = w0.copy()
    
# compute initial error
errs = []
S = 1. * ((X @ w) > 0)
err = np.abs(S - y).sum()
errs += [err]

# run training
for ep in range(num_epochs):
    S = 1. * ((X @ w) > 0)
    delta = y - S
    w += lr * (delta[:,None] * X).sum(0)
    err = np.abs(delta).sum()
    errs += [err]
    
    if ep % print_every == 0:
        print(f"ep: {ep} err: {err}")
    
    if err == 0:
        print("solution found -> EXITING!")
        break
        
plt.plot(errs, '.-');
plt.ylabel("errors");
```

### Perceptron capacity

Let's first wrap the training procedures into a function

```python
def train_perceptron(X, y, learn_bias = True, lr = 0.1, num_epochs = 1000,
                     print_every = 10, parallel = False, verbose = False, renormalize = False):
    
    # init weight vector
    N = X.shape[-1]
    w = np.random.randn(N)# / np.sqrt(N)
    if learn_bias:
        w = np.r_[w, np.random.randn()]

    # compute initial error
    errs = []
    S = 1. * ((X @ w) > 0)
    err = np.abs(S - y).sum()
    errs += [err]

    for ep in range(1,num_epochs+1):
        if parallel:
            S = 1. * ((X @ w) > 0)
            delta = y - S
            w += lr * (delta[:,None] * X).sum(0)
            if renormalize:
                w *= np.sqrt(N / (w**2).sum())
        else:
            perm = np.random.permutation(P)
            for mu in perm:
                s = 1. * ((w @ X[mu]) > 0)
                deltamu = y[mu] - s
                if deltamu != 0:
                    w += lr * deltamu * X[mu]
                if renormalize:
                    w *= np.sqrt(N / (w**2).sum())
        S = 1. * ((X @ w) > 0)
        delta = y - S
        err = np.abs(delta).sum()
        errs += [err]

        if verbose and ep % print_every == 0:
            print(f"ep: {ep} err: {err}")

        if err == 0:
            if verbose:
                print("solution found -> EXITING!")
            break

        
    return w, err, errs
```

and then span over multiple values of α

```python
# set parameters
lr = 0.01
num_epochs = 1000
print_every = 1e10
parallel = False
learn_bias = False #Already defined in the beginning, should run from beginning if you want to include it (X changes accordingly)

Ns = [100]
alphas = np.arange(0.5, 2.5, 0.2) # coarse
# alphas = np.arange(1.6, 2.4, 0.1) # zoom around critical α
samples = 1

err_span = np.zeros((len(Ns), len(alphas), samples))

for sample in range(samples):
    for iN, N in enumerate(Ns):
        print(f"doing N {N} sample {sample}")
        for ialpha, alpha in enumerate(alphas):

            # generate input-output task
            P = int(alpha * N)
            X = np.random.randn(P, N)
            y = np.random.randint(2, size=P)
            if learn_bias:
                X = np.c_[X, np.ones((P,1))]

            _, err, _ = train_perceptron(X, y, learn_bias=learn_bias, lr=lr, num_epochs=num_epochs,
                                      print_every=print_every, parallel=parallel)
            err_span[iN, ialpha, sample] = err
```

```python
# watch single run
# sample = 0
# plt.plot(alphas, err_span[:,:,sample].T, '.-');

# watch averages
for iN, N in enumerate(Ns):
    plt.errorbar(alphas, err_span[iN].mean(-1), yerr=err_span[iN].std(-1), label=str(N));
plt.xlabel('α')
plt.ylabel('err');
plt.legend();
```

### Train using Hebbian learning

```python
E = w@X[mu] * y[mu]
E<0
```

```python
def train_perceptron_Hebbian(X, y, learn_bias = True, lr = 0.1, num_epochs = 1000,
                     print_every = 10, parallel = False, verbose = False, renormalize = False):
    
    # init weight vector
    N = X.shape[-1]
    w = np.random.randn(N)# / np.sqrt(N)
    if learn_bias:
        w = np.r_[w, np.random.randn()]

    # compute initial error
    errs = []
    S = 1. * ((X @ w) > 0)
    err = np.abs(S - y).sum()
    errs += [err]

    for ep in range(1,num_epochs+1):
        if parallel:
            S = 1. * ((X @ w) > 0)
            delta = y - S
            w += lr * (delta[:,None] * X).sum(0)
            if renormalize:
                w *= np.sqrt(N / (w**2).sum())
        else:
            perm = np.random.permutation(P)
            for mu in perm:
                # Compute E
                E = w@X[mu] * y[mu]
                s = 1. * ((E) <= 0)
                deltamu = s*X[mu]*y[mu]
                if deltamu != 0:
                    w += lr * deltamu
                if renormalize:
                    w *= np.sqrt(N / (w**2).sum())
        S = 1. * ((X @ w) > 0)
        delta = y - S
        err = np.abs(delta).sum()
        errs += [err]

        if verbose and ep % print_every == 0:
            print(f"ep: {ep} err: {err}")

        if err == 0:
            if verbose:
                print("solution found -> EXITING!")
            break

        
    return w, err, errs
```

```python
# set parameters
lr = 0.01
num_epochs = 1000
print_every = 1e10
parallel = False
learn_bias = False #Already defined in the beginning, should run from beginning if you want to include it (X changes accordingly)

Ns = [100]
alphas = np.arange(0.5, 3, 0.2) # coarse
# alphas = np.arange(1.6, 2.4, 0.1) # zoom around critical α
samples = 1

err_span = np.zeros((len(Ns), len(alphas), samples))

for sample in range(samples):
    for iN, N in enumerate(Ns):
        print(f"doing N {N} sample {sample}")
        for ialpha, alpha in enumerate(alphas):

            # generate input-output task
            P = int(alpha * N)
            X = np.random.randn(P, N)
            y = 2*np.random.randint(2, size=P)-1
            if learn_bias:
                X = np.c_[X, np.ones((P,1))]

            _, err, _ = train_perceptron_Hebbian(X, y, learn_bias=learn_bias, lr=lr, num_epochs=num_epochs,
                                      print_every=print_every, parallel=parallel)
            err_span[iN, ialpha, sample] = err
```

```python
# watch single run
# sample = 0
# plt.plot(alphas, err_span[:,:,sample].T, '.-');

# watch averages
for iN, N in enumerate(Ns):
    plt.errorbar(alphas, err_span[iN].mean(-1), yerr=err_span[iN].std(-1), label=str(N));
plt.xlabel('α')
plt.ylabel('err');
plt.legend();
```

### Teacher-student perceptron and generalization: theory

```python
# Define a bunch of useful function

math_prefix = np
# math_prefix = math

log = math_prefix.log
exp = math_prefix.exp
sqrt = math_prefix.sqrt
PI = math_prefix.pi
atan = math.atan
asin = math.asin
acos = math.acos
sign = np.sign

def G(x):
    return 1/sqrt(2*PI) * exp(-0.5 * x**2)

def erfsq2(x):
    return erf(x / sqrt(2))

def H(x):
    if x < 5:
        return 0.5 * (1 - erfsq2(x))
    else:
        return G(x) * (1./x - 1./x**3 + 3./x**5)
    
def IntR(t, R):
    return exp(-0.5 * R * t**2) / H(-sqrt(R) * t) * multivariate_normal.pdf(t, mean=0, cov=1)

def IR(R):
    xmin, xmax = -10., 10.
    return cubature(IntR, 1, 1, [xmin], [xmax], args=(R,))[0][0]

def eqR(R, α = 1):
    return R / sqrt(1-R) - α / PI * IR(R)
```

```python
# solve the theory spanning over α

αs = np.arange(1, 30, 1)
Rs = np.zeros(len(αs))
gen_errors = np.zeros(len(αs))

R0 = 0.3
for iα, α in enumerate(αs):
    R = optimize.fsolve(eqR, R0, args=(α)).item()
    gen_errors[iα] = 1/PI * acos(R)
    Rs[iα] = R
    R0 = R
```

```python
plt.plot(αs, Rs, '--', lw=5, label="W * T / N");
plt.plot(αs, gen_errors, lw=5, label="gen error");
plt.legend();
plt.xlabel('α');
plt.ylabel('overlap / gen error');
# plt.rc('font', size=20);
```

### Teacher-student perceptron and generalization: practice

```python
# generate teacher vector and test patterns
N = 500
Ptest = 10000
T = np.random.randn(N)
T *= np.sqrt(N / (T**2).sum())
Xtest = np.random.randn(Ptest, N)
ytest_sign = np.sign(Xtest @ T)
```

```python
# set parameters
lr = 0.001
num_epochs = 1000
print_every = 1e10
parallel = True
renormalize = False
learn_bias = False

α_span = np.arange(1, 30, 1)
w_span = np.zeros((len(α_span), N))
normw_span = np.zeros(len(α_span))
err_span = np.zeros(len(α_span))
ov_span = np.zeros(len(α_span))
gen_err_span = np.zeros(len(α_span))

for iα, α in enumerate(α_span):

    print(f"doing α: {α}")
    
    P = int(α * N)
    X = np.random.randn(P, N)
    y = 1. * (X @ T > 0)

    w, err, errs = train_perceptron(X, y, learn_bias=learn_bias, lr=lr, num_epochs=num_epochs,
                              print_every=print_every, parallel=parallel, renormalize=renormalize)
    w_span[iα] = w
    normw_span[iα] = (w**2).sum()
    err_span[iα] = err
    ov_span[iα] = w @ T / N
    gen_err_span[iα] = 1 - (np.sign(Xtest @ w) * ytest_sign > 0).sum() / Ptest
```

#### Run with renormalize = False

```python
# plt.plot(α_span, err_span, color='black', label='train error');

plt.plot(α_span, ov_span, '.', color='blue', label="W * T / N")
plt.plot(αs, Rs, '--', color='blue', label="W * T / N theory");

plt.plot(α_span, gen_err_span, '.', color='red', label="gen error");
plt.plot(αs, gen_errors, color='red', label="gen error theory");

plt.legend();
plt.xlabel('α');
plt.ylabel('overlap / gen error');
```
